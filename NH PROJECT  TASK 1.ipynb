__DETECTING OUTLIERS USING IQR METHOD__

```python
import numpy as np

data = np.random.normal(loc=0, scale=1, size=1000)

# data
# calculate first quartile 25th percentile

Q1 = np.percentile(data, 25)
# Q1 = np.percentile()
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1
# IQR

# Define lower bound and upper bound

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify the outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("The outliers using IQR method:", outliers)
```

__REMOVING OUTLIERS USING IMPUTATION METHOD__

```python
out1 = outliers

# calculate the mean of non-outliers values
non_outliers_mean = np.mean([x for x in data if x not in out1])
# non_outliers_mean

# Impute outliers with mean of the outliers values
imputed_data = [non_outliers_mean if x in out1 else x for x in data]
```

__Dimensionality Reduction – perform a principal component analysis to reduce the dimensions of your 
data and provide a useful interpretation of the results (Provide your interpretation in four (4) bullet points maximum)__

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
quan_var=df1.select_dtypes(include=['int64' ,'float64'])
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(quan_var)
```

```python
pca=PCA(n_components=2)
principal_components=pca.fit_transform(scaled_data)
pca_df=pd.DataFrame(data=principal_components,columns=['PC1','PC2'])
explained_variance=pca.explained_variance_ratio_
print('Explained variance by component:',explained_variance)
pca_df.head()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(pca_df['PC1'],pca_df['PC2'])
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
